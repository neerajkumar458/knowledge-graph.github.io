```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/HP-DeskJet-2723-Wireless-Printer/product-reviews/B08D9MNH4B/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
HEADERs = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36 Edg/113.0.1774.50', 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage = requests.get(URL, headers=HEADERS)
```

```python
type(webpage.content)
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name' )
```

```python
names
```

```python
customer_name=[]
for i in range(0,len(names)):
    customer_name.append(names[i].get_text())
```

```python
title = soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:]=[titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:]=[titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate=[]
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content=[]
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:]=[reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
review_content[:]=[reviews.rstrip('\n') for reviews in review_content]
review_content
```
